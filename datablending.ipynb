```python
from os import listdir
from os.path import isfile, join
import csv
import pandas as pd
```

```python
mypath = '<directory of summaries>'
summaries = [summary for summary in listdir(mypath) if isfile(join(mypath, summary))]
```

```python
summaries = [summary for summary in summaries if summary.startswith('en')]
```

```python
summaries
```

```python
finallist = []
rowlist = []
for summary in summaries:
    filename = mypath + summary
    with open(filename, 'rb') as csv_in:
        myreader = csv.reader(csv_in, delimiter = ',')
        next(myreader)
        
        for row in myreader:
            country = row[0]
            if country not in ['',' World']:
                for i in range(15):
                    value = row[i].strip()
                    if i >= 5 and value != '':
                        value = float(value)
                    elif i >= 5 and value == '':
                        value = 0.0

                    if i == 4:
                        if row[i] == 'Export(US$ Mil)':
                            value = 'Exports (in US$ Mil)'
                        elif row[i] == 'Import(US$ Mil)':
                            value = 'Imports (in US$ Mil)'

                    rowlist.append(value)
                finallist.append(rowlist)
                rowlist = []
            
    
    if country not in ['',' World']:
        print 'Saved:', country, '-', summary
                
```

```python
labels = []   
for summary in summaries:
    filename = mypath + summary
    with open(filename, 'rb') as csv_in:
        myreader = csv.reader(csv_in, delimiter = ',')
        
        for row in myreader:
            for i in range(15):
                labels.append(row[i])
            break
    break
```

```python
labels
```

```python
df = pd.DataFrame.from_records(finallist, columns = labels)
```

```python
df
```

```python
df.to_csv('all_summaries.csv', index = False)
```

```python
melted = pd.melt(df, id_vars = labels[:5], var_name = 'Years')
```

```python
melted
```

```python
melted.to_csv('all_summaries_melted.csv', index = False)
```
